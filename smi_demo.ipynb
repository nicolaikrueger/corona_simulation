```python
import matplotlib.pyplot as plt
import numpy as np
import random
import time

#epedemie parameters
t = 1 #current day 
i = 16 #total amount of infected people
infection_rate = 1.395#infection rate; new infected people by one infected person
i_new = 0 #novel invections within a day
i_new_yesterday = 1 #variable to hold the amount of infected people of the last day simulated
iterations = 30 #days the simulation shall run
simulation_iterations = [] #list holding the iteration index completely
simulation_infections = [] #list holding the infection sum completely

def simulation(iterations, t, i, i_new, i_new_yesterday):  
    while t < iterations:
        print ("===== ITERATION DAY ", t, "=====")
        
        i_new = infection_rate*i_new_yesterday
        i_new_yesterday = i_new
        print ("Newly infected people: ", i_new)
        
        i = i + i_new
        print ("Total Corona Sum: ", i)
        
        simulation_iterations.append(t)
        simulation_infections.append(i)
        
        t = t + 1 

        print (" ")
        
simulation(iterations, t, i, i_new, i_new_yesterday) 
plt.plot(simulation_iterations, simulation_infections, label="Infections")
plt.xlabel('Day')
plt.ylabel('Corona Infections')
plt.title('Corona Simulation')
plt.grid(True)
plt.show()
```
